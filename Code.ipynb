# Spilitting of Data in Train And Test

```python
# input Variable
X=df3.drop(['target'],axis=1)
y=df3['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2020)
# spilitting of data in X_train,X_test,y_train,y_test
```

# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(X_train)
X_train_ss=ss.transform(X_train)
X_test_ss=ss.transform(X_test)
# we use feature saclling because some algo work well on that data..beacuse the units are different of all the features
#we need to feature scale all that features with respect to (0,1)
```

# Apply Some Machine Learning Models And See The score

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

# SVM Classifier

```python
from sklearn.svm import SVC
support_vector_classifier=SVC(kernel='rbf')# we use that in bydeafault parameter
support_vector_classifier.fit(X_train,y_train)
```

```python
support_vector_classifier.score(X_test,y_test)
```

```python
#if we need accuracy score then:
y_predict_svm=support_vector_classifier.predict(X_test)
accuracy_score(y_test,y_predict_svm)
```

```python
#if then then we give standard scale data
support_vector_classifier_stand=SVC(kernel='rbf')# we use that in bydeafault parameter
support_vector_classifier_stand.fit(X_train_ss,y_train)
```

```python
#if we need accuracy score then:
y_predict_svm_sc=support_vector_classifier_stand.predict(X_test_ss)
accuracy_score(y_test,y_predict_svm_sc)
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knnc=KNeighborsClassifier(n_neighbors=5,p=2)
```

```python
knnc.fit(X_train,y_train)
```

```python
#if we need accuracy score then:
y_predict_knn=knnc.predict(X_test)
accuracy_score(y_test,y_predict_knn)
```

```python
#give standard scale data
knnc_stand=KNeighborsClassifier(n_neighbors=5,p=2)
knnc_stand.fit(X_train_ss,y_train)
```

```python
#if we need accuracy score then:
y_predict_knn_sc=knnc_stand.predict(X_test_ss)
accuracy_score(y_test,y_predict_knn_sc)
```

# Logictics Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
y_pre_log=lr.predict(X_test)
accuracy_score(y_test,y_pre_log)
```

```python
#if we give standard scale data
lr_stand=LogisticRegression()
lr_stand.fit(X_train_ss,y_train)
y_pre_log_sc=lr_stand.predict(X_test_ss)
accuracy_score(y_test,y_pre_log_sc)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier(criterion='gini')
dtc.fit(X_train,y_train)
y_pre_dis=dtc.predict(X_test)
accuracy_score(y_test,y_pre_dis)
```

```python
#if we give Feature scale Data
dtc_stand=DecisionTreeClassifier(criterion='gini')
dtc_stand.fit(X_train_ss,y_train)
y_pre_dis_sc=dtc_stand.predict(X_test_ss)
accuracy_score(y_test,y_pre_dis_sc)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier( n_estimators=20,criterion='gini')
rfc.fit(X_train,y_train)
y_pre_rn=rfc.predict(X_test)
accuracy_score(y_test,y_pre_rn)
```

```python
#if we give standard scale data
rfc_stand=RandomForestClassifier( n_estimators=100,criterion='gini')
rfc_stand.fit(X_train_ss,y_train)
y_pre_rn_stand=rfc_stand.predict(X_test_ss)
accuracy_score(y_test,y_pre_rn_stand)
```
